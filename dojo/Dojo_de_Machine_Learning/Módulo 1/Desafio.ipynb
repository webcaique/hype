# Tarefa 5

examine o Dataframe, trate valores nulos e linhas duplicadas.

```python
dataFrame.isnull().sum()
```

```python
#Função para calcular mediana de strings
def list_function(column):
  nums = list()
  for number in dataFrame[column]:
    if(str(number)[0] == "$"):
      nums.append(str(number)[1:])
    else:
      nums.append(str(number)[:-1])
  return nums


for column in dataFrame.columns:
  pos = 0
  sampleColumn =  dataFrame[column].sample(5).to_list()
  for key, sample in enumerate(sampleColumn):
    if(str(sample) != "nan"):
      pos = key
      break
  try:
    strSample = str(sampleColumn[pos])
    if("%" not in strSample and "$" not in strSample  and not strSample[0].isnumeric()):
      dataFrame[column] = dataFrame[column].fillna("NA")
    else:
      dataFrame[column] = dataFrame[column].fillna(dataFrame[column].mean)
  except:
    text = f"${list_function(column).mean}"
    if(sampleColumn[pos][0]!="$"):
      text = f"{list_function(column).mean}%"
    dataFrame[column] = dataFrame[column].fillna(text)
```
