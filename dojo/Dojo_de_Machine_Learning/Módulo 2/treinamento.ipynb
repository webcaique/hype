```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataFrame = pd.read_csv(arqPath) # lê um arquivo csv
display(dataFrame) # mostra esses dados
```

```python
#GRÁFICO DE DISPERSÃO
fig = sns.scatterplot(data=dataFrame, x="Age", y="MaxHR", hue="ExerciseAngina", style="Sex", palette = ['green', '#ff0000'])
#hue = fazer a separação dos dados por cor em relação à uma coluna
#style = muda o estilo dos pontos dependendo da coluna
#palette = paleta de cor que se aplica para o hue
plt.show(fig)
```

```python
#GRÁFICO DE DISPERSÃO RELACIONAL
gfRel = sns.relplot(data=dataFrame, x="Age", y="MaxHR", hue='Sex', col='ExerciseAngina')
#col = separa em dois gráficos dependendo da coluna escolhida
gfRel.set_titles("Faz exercícios: {col_name}") # muda o título de uma coluna
plt.show(gfRel)
```

```python
#GRÁFICO DE LINHAS
grafLin = sns.lineplot(data=dataFrame, x="RestingBP",y="Cholesterol")
plt.show(grafLin)
```

```python
#HISTOGRAMA
hist = sns.displot(data=dataFrame,x="Age", kind="hist", hue="ChestPainType", col="Sex", kde=True)
#kde = ativa a linha da dispersão
#kind = muda o tipo de gráfico
plt.show(hist)
```

```python
#REGRESSÃO LINEAR
grafReg = sns.regplot(data=dataFrame, x="Age", y="MaxHR")
plt.show(grafReg)
```

```python
#Gráfico de regressão com a linha que indica a direção
grafReg = sns.lmplot(data=dataFrame, x="Age", y="MaxHR", hue="Sex", markers=["o","x"])
plt.show(grafReg)
```

```python
#Gráfico que junta a dispersão com a indicação das densidados em certas áreas
joinPlot = sns.jointplot(data=dataFrame, x="Age", y="MaxHR", hue="Sex")
plt.show(joinPlot)
```

```python
joinPlot = sns.jointplot(data=dataFrame, x="Age", y="MaxHR", hue="Sex", kind="kde")
plt.show(joinPlot)
```

```python
joinPlot = sns.jointplot(data=dataFrame, x="Age", y="MaxHR", kind="reg")
plt.show(joinPlot)
```

```python
joinPlot = sns.jointplot(data=dataFrame, x="Age", y="MaxHR", hue="Sex", kind="hist")
plt.show(joinPlot)
```

```python
joinPlot = sns.jointplot(data=dataFrame, x="Age", y="MaxHR", kind="hex")
plt.show(joinPlot)
```

```python
#Mostra diversas dispersões e gráficos
pairPlot = sns.pairplot(data=dataFrame, hue="Sex")
plt.show(pairPlot)
```

```python
plt.show(sns.violinplot(data=dataFrame, x="Sex", y="MaxHR"))
```

```python
#Faz um gráfico de calor em relação a correlação dos dados
#annot = exibe ou não as porcentagens das correlação
plt.show(sns.heatmap(dataFrame.corr(), annot=True))
```

```python
p = sns.FacetGrid(dataFrame, col="Sex", row='ChestPainType')
p_map = p.map(sns.histplot, "Age")
plt.show(p_map)
```

```python
dados = {
    'X': [1,2,3,4,5,6],
    'Y1': [120,110,130,145,118,125],
    'Y2': [95,54,86,77,90,81]
}

base = pd.DataFrame(dados)
```

```python
fig, ax = plt.subplots(figsize=(12,7))

ax.plot(base.X, base.Y1, label='Y1')

for i in range(0,len(base.X)):
  ax.annotate(base.Y1[i], (base.X[i], base.Y1[i]))

ax.set_yticks([0,160])
ax.yaxis.set_visible(False)
ax.spines[['top','right','left']].set_visible(False)


plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,7))

ax.plot(base.X[0:4], base.Y1[0:4], label='Y1')
ax.plot(base.X[3:], base.Y1[3:], 'r--')
for i in range(0,len(base.X)):
  ax.annotate(base.Y1[i], (base.X[i], base.Y1[i]))

ax.set_yticks([0,160])
ax.yaxis.set_visible(False)
ax.spines[['top','right','left']].set_visible(False)


plt.show()
```
